### Document Loader

##### Text Loader

```python
from langchain_community.document_loaders.text import TextLoader

loader = TextLoader("../data/sample.txt")
text_document = loader.load()

print(text_document[0])
print("========================")
print(text_document[0].metadata)
print("========================")
print(text_document[0].page_content)
```

#### CSV Loader

```python
# Get specific cloumns in csv file using CSV loader
from langchain_community.document_loaders.csv_loader import CSVLoader

loader = CSVLoader(file_path='../data/sample.csv')
docs = loader.load()
print(docs)
# If you want to extract specific columns from the loaded docs:
selected_columns = [ 'First Name', 'Last Name']
filtered_docs = []
for doc in docs:
    filtered_doc = {col: doc.metadata[col] for col in selected_columns if col in doc.metadata}
    filtered_docs.append(filtered_doc)

print(filtered_docs)
```

#### Web Based Loader

```python
from langchain_community.document_loaders import WebBaseLoader
import bs4

loader = WebBaseLoader("https://www.wikipedia.org/")
web_document = loader.load()
print(web_document)
```

```python
# loading specific tags with class potd-recent, external text
tags_loader = WebBaseLoader(web_paths=("https://www.wikipedia.org/",), bs_kwargs=dict(parse_only=bs4.SoupStrainer(class_=("footer-sidebar", "external text"))))
tags_document = tags_loader.load()
print(tags_document)
```
